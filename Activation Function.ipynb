```python
import math
import matplotlib.pyplot as plt
import numpy as np
x= np.arange(-5,5,0.5)
print(x)
```

```python
def Linear_Activation_Function(x):
    a=[]
    for i in x:
        a.append(i)
    return a
#i is array element
```

```python
y=Linear_Activation_Function(x)
plt.plot(x,y)
plt.grid()
```

```python
def Relu_Activation_Function(x):
  a=[]
  for i in x:
    if( i > 0):
      a.append(i)
    else:
      a.append(0)
  return a
```

```python
y1=Relu_Activation_Function(x)
plt.plot(x,y1)
plt.grid()
```

```python
def Sigmoid_Activation_Function(x):
 a=[]
 for i in x:
    a.append(1/(1+np.exp(-i)))
 return a
```

```python
y2=Sigmoid_Activation_Function(x)
plt.plot(x,y2)
plt.grid()
```

```python
def tanh_Activation_Function(x):
  a=[]
  for i in x:
    a.append((np.exp(i)-np.exp(-i))/(np.exp(i)+np.exp(-i)))
    
  return a
```

```python
y3=tanh_Activation_Function(x)
plt.plot(x,y3)
plt.grid()
```

```python
#f1 = x**3 + x**2 + x + 2
def Functional_Activation_Function(x):
  a=[]
  for i in x:
    a.append(i**3 + i**2 + i + 2) 
  return a
```

```python
y4=Functional_Activation_Function(x)
plt.plot(x,y4)
plt.grid()
```
